<h2 style="color:#000047; margin-top:0;">Instrucciones de la Tarea</h2>
<ol style="font-size:1.08em;">
<li>Clona el repositorio: <a href="https://github.com/gdesirena/HyAIA_Repositorio_Clase" target="_blank"><b>HyAIA_Repositorio_Clase</b></a></li>
<li>Crea una nueva branch con el formato <code>[Nombre_Apellido_Doc_Clase]</code></li>
<li>Agrega la siguiente estructura de carpetas y archivos:
  <ul>
    <li><b>Documentos</b> <span style='color:#888;'>(agrega algún archivo: txt, pdf, etc.)</span></li>
    <li><b>Practicas</b> <span style='color:#888;'>(agrega las prácticas realizadas)</span></li>
    <li><b>Tareas</b> <span style='color:#888;'>(agrega las tareas realizadas)</span></li>
    <li><b>Proyecto Final</b></li>
  </ul>
</li>
<li>En la raíz del repositorio agrega los archivos <code>readme.md</code> y <code>.gitignore</code></li>
<li>Agrega todos los archivos al repositorio:<br>
<code>git add --all</code></li>
<li>Realiza un commit con un mensaje descriptivo:<br>
<code>git commit -m "Comentario del commit"</code></li>
<li>Haz push de tus cambios:<br>
<code>git push</code></li>
</ol>
